```python
import cv2
import matplotlib.pyplot as plt 
%matplotlib inline 
import numpy as np 
```

```python
source = cv2.imread("img_05887.jpg")
template = cv2.imread("shark_copy.jpg")
```

```python
(tempH, tempW) = template.shape[:2]
print (tempH, tempW)
```

```python
plt.imshow(source)
plt.show()
```

```python
result = cv2.matchTemplate(source, template, cv2.TM_CCOEFF)
```

```python
result
```

```python
(minVal, maxVal, minLoc, (x,y)) = cv2.minMaxLoc(result)
```

```python
# Bouding box on the source image 
cv2.rectangle(source, (x,y), (x+tempW,y+tempH),(0,255,0), 2)
plt.imshow(source)
```
